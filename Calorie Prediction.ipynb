```python
from sklearn.model_selection import train_test_split
```

```python
df[df['Duration']<10][df['Calories']>100]
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(df['Duration'],df['Calories'])
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df = df[df['Calories']<250]
print(df.head())
print(df.describe())
```

```python
X = df.drop(['User_ID','Calories'],axis=1)
y = df['Calories']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print("Mean absolute error is ",mean_absolute_error(lr_pred,y_test))
print("Root mean squared error is",np.sqrt(mean_squared_error(lr_pred,y_test)))
```
